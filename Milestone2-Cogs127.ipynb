```python
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
df = pd.read_csv('ZoomSurvey.csv')
df = df.drop('Timestamp', 1)
df
```

```python
df = df.rename(columns={'What is your age and occupation? ': 'Age','What calendar do you use for zoom links? If you don\'t, why not?': 'Calendar use', 'If there was a native zoom calendar, would you use it?': 'Zoom Calendar','If you do use a calendar for zoom links, how would you rate your experience? (1 = awful, 5 = perfect)': 'Experience Rating','Do you have a place you store recurring zoom meeting links?':'Storage', 'How many average hours are you on zoom per week?': 'Avg Time per Week (Hours)'})
df
```

```python
df.loc[:,"Age"]
#Since all of our users are students so far, I'm going to remove student from the category and just have it as age
```

```python
df.Age = df.Age.str.extract('(\d+)')
df
```

```python
data = df['Avg Time per Week (Hours)']

x1 = data
y1 = range(len(data))

plt.figure(figsize=(20, 9))
plt.bar(x1, y1)
plt.ylabel("Frequency")
plt.xlabel("Avg Time Per Week on Zoom (Hours)")
plt.show()
```

```python
age_data = df['Age']
age_data = age_data.sort_values(ascending=True)

x2 = age_data
y2 = range(len(age_data))

plt.figure(figsize=(10, 9))
plt.bar(x2, y2)
plt.ylabel("Frequency")
plt.xlabel("Age of Survey Responders")
plt.show()
```

```python
import numpy as np
#experience rating histogram


#getting rid of NaN values so we can plot
ratings = df['Experience Rating']
ratings = [x for x in ratings if np.isnan(x) == False]


x3 = ratings
y3 = range(len(ratings))

plt.figure(figsize=(10, 9))
plt.bar(x3, y3)
plt.ylabel("Frequency")
plt.xlabel("User Ratings of Zoom Calendar System")
plt.show()
```

```python
zoom_cal = df['Zoom Calendar']

x4 = zoom_cal
y4 = range(len(zoom_cal))

plt.figure(figsize=(10, 9))
plt.bar(x4, y4)
plt.ylabel("Frequency")
plt.xlabel("Native Zoom Calendar Approval")
plt.show()
```
